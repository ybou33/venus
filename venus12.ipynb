```python
import math
import pandas as pd
from pandas import DataFrame, read_csv

#récupération des données
file = r'venus.xlsx'
df = pd.read_excel(file)
t=df['t']
x=df['x']
y=df['y']
vx=['rien']
for i in range(1,19):
    vx.append((x[i+1]-x[i-1])/(t[i+1]-t[i-1]))

#vitesse
vy=['rien']
for i in range(1,19):
    vy.append((y[i+1]-y[i-1])/(t[i+1]-t[i-1]))

v=['rien']
for i in range(1,19):
    v.append(int(math.sqrt(vx[i]**2+vy[i]**2)))
    ax=['rien']
ax.append('rien')
for i in range(2,18):
    ax.append((vx[i+1]-vx[i-1])/(t[i+1]-t[i-1]))

#accélération
ay=['rien','rien']
for i in range(2,18):
    ay.append((vy[i+1]-vy[i-1])/(t[i+1]-t[i-1]))
a=['rien','rien']
for i in range(2,18):
    a.append(math.sqrt(ax[i]**2+ay[i]**2))

#graphique
import matplotlib.pyplot as plt
plt.scatter(x,y)
axee=plt.axes()

for i in range(2,18):
    bx=5E12*ax[i]
    by=5E12*ay[i]

    #q = axee.quiver(x[i],y[i],bx,by,units='xy' ,scale=0.2)
    r = axee.quiver(x[i],y[i],bx,by,units='xy' ,scale=1,color='g')
plt.axis('equal')
plt.show()
plt.scatter(x,y)
axee=plt.axes()

for i in range(2,18):
    bx=1000000*vx[i]
    by=1000000*vy[i]

    #q = axee.quiver(x[i],y[i],bx,by,units='xy' ,scale=0.2)
    r=axee.quiver(x[i],y[i],bx,by,units='xy' ,scale=1)
plt.axis('equal')
plt.show()
plt.scatter(x,y)
axee=plt.axes()


bx=3000000*vx[10]
by=3000000*vy[10]
cx=3000000*vx[8]
cy=3000000*vy[8]

axee.quiver(x[9],y[9],bx,by,units='xy' ,scale=1,color='g')
axee.quiver(x[9],y[9],cx,cy, units='xy' ,scale=1,color='b')
axee.quiver(x[9],y[9],-cx+bx,-cy+by,units='xy' ,scale=1,color='r')
plt.axis('equal')
plt.show()
```
